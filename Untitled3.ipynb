```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
print("Loading dataset...")
data = pd.read_csv('spam.csv', encoding='latin-1')
data = data.rename(columns={"v1": "label", "v2": "text"})  # Adjust column names if needed
data = data[['label', 'text']]  # Keep only the label and text columns

# Step 2: Data Exploration
print("\nDataset head:")
print(data.head())  # Show first few rows of the data

# Check for missing values
print("\nMissing values check:")
print(data.isnull().sum())  # No missing values should be expected

# Step 3: Encode labels: 'spam' as 1, 'ham' as 0
data['label'] = data['label'].map({'spam': 1, 'ham': 0})
print("\nEncoded labels:")
print(data['label'].value_counts())  # Check the class distribution

# Step 4: Data Preprocessing
# Additional steps: Removing duplicates and resetting the index
data = data.drop_duplicates()  # Drop duplicate rows
data = data.reset_index(drop=True)  # Reset the index after dropping rows

# Step 5: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(data['text'], data['label'], test_size=0.2, random_state=42)
print("\nTraining data size:", X_train.shape)
print("Testing data size:", X_test.shape)

# Step 6: Text Vectorization using TF-IDF
print("\nVectorizing text data...")
vectorizer = TfidfVectorizer(min_df=5, max_df=0.7, stop_words='english', ngram_range=(1, 2))  # Use 1-grams and 2-grams
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Display the shape of the vectorized data
print("\nVectorized training data shape:", X_train_tfidf.shape)
print("Vectorized test data shape:", X_test_tfidf.shape)

# Step 7: Train Logistic Regression Model
print("\nTraining Logistic Regression model...")
model = LogisticRegression(max_iter=1000, random_state=42)  # Increase max_iter if convergence warning occurs
model.fit(X_train_tfidf, y_train)

# Step 8: Make Predictions
print("\nMaking predictions on test data...")
y_pred = model.predict(X_test_tfidf)

# Step 9: Evaluate the Model
print("\nEvaluating the model...")

# 1. Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy: {accuracy:.4f}")

# 2. Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(conf_matrix)

# 3. Classification Report
class_report = classification_report(y_test, y_pred)
print("\nClassification Report:")
print(class_report)

# Step 10: Plotting the Confusion Matrix
print("\nPlotting Confusion Matrix...")
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=["Ham", "Spam"], yticklabels=["Ham", "Spam"])
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# Step 11: Test the Model with New Input
print("\nTesting the model with a new message...")
new_message = ["Congratulations! You've won a $1,000 gift card. Click here to claim."]
new_message_tfidf = vectorizer.transform(new_message)
prediction = model.predict(new_message_tfidf)

# 1. Displaying the result
if prediction == 1:
    print("\nPrediction: Spam!")
else:
    print("\nPrediction: Not Spam (Ham)")

# Step 12: Model Performance Insights
print("\nModel Insights:")
# Additional: Look at the coefficients of the Logistic Regression model to gain insight into the most important features
coefficients = model.coef_.flatten()  # No need for .toarray() as coef_ is already a NumPy array
features = vectorizer.get_feature_names_out()
feature_importance = pd.DataFrame({"feature": features, "coefficient": coefficients})
feature_importance = feature_importance.sort_values(by="coefficient", ascending=False)

print("\nTop 10 most important features (keywords) for predicting Spam messages:")
print(feature_importance.head(10))

# Step 13: Save the Model for Future Use
import joblib
print("\nSaving the model and vectorizer for future use...")
joblib.dump(model, 'spam_classifier_model.pkl')
joblib.dump(vectorizer, 'tfidf_vectorizer.pkl')

print("\nModel and vectorizer saved successfully!")

```
